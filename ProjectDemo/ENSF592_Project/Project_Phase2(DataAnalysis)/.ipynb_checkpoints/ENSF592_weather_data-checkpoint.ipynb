```python
import pandas as pd
import numpy as np
```

```python
traffic_incidents = pd.read_csv("Traffic_Incidents.csv")
traffic_incidents['START_DT'] = pd.to_datetime(traffic_incidents['START_DT'])
traffic_incidents_2018 = traffic_incidents.loc[traffic_incidents["START_DT"].dt.year == 2018, :]

traffic_incidents_2018 = traffic_incidents_2018.rename(columns = {'START_DT' : 'Date/Time'}) 

weather_2018 = pd.read_csv("https://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=50430&Year=2018&Month=1&Day=1&timeframe=2&submit=Download+Data")
weather_vis_2018 = pd.read_csv("https://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=50430&Year=2018&Month=1&Day=1&timeframe=1&submit=Download+Data")

weather_2018 = weather_2018[['Date/Time','Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)']]

weather_vis_2018 = weather_vis_2018[['Date/Time', 'Temp (°C)', 'Visibility (km)', 'Weather']]
display(traffic_incidents_2018, weather_2018, weather_vis_2018)
```
